```python
id_list = os.listdir(se_mask_attn_path)
```

```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

# ---------------------------
# Existing variables you already have
# ---------------------------
ID_IDX = 0
B_IDX = 15

b_list = os.listdir(se_mask_attn_path + "/" + id_list[ID_IDX])

se_mask_imgs_path = Path(se_mask_attn_path) / id_list[ID_IDX] / b_list[B_IDX]
mask_imgs_path = Path(mask_attn_path) / id_list[ID_IDX] / b_list[B_IDX]

print(se_mask_imgs_path)
print(mask_imgs_path)

# ---------------------------
# Load images 00.png to 07.png
# ---------------------------
img_names = [f"{i:02d}.png" for i in range(8)]  # ["00.png", ..., "07.png"]

se_mask_imgs = []
mask_imgs = []

for name in img_names:
    se_mask_imgs.append(Image.open(se_mask_imgs_path / name))
    mask_imgs.append(Image.open(mask_imgs_path / name))

# ---------------------------
# Plot: top row = se_mask, bottom row = mask
# ---------------------------
n = len(img_names)
fig, axes = plt.subplots(2, n, figsize=(3*n, 6))

for i, img in enumerate(mask_imgs):
    axes[0][i].imshow(img)
    axes[0][i].axis("off")
    #axes[0][i].set_title(f"spatial Block {i}")

for i, img in enumerate(se_mask_imgs):
    axes[1][i].imshow(img)
    axes[1][i].axis("off")
    #axes[1][i].set_title(f"channel+spatial Block {i}")

plt.tight_layout()
plt.show()
```
